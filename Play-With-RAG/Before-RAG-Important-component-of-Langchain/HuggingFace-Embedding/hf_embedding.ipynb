```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
embeddings
```

```python
text = "hello sir how are you hope you are if not why?"
```

```python
query_result = embeddings.embed_query(text=text)
query_result
```

```python
len(query_result)
```

```python
doc_result = embeddings.embed_documents([text, "This is another test document."])
doc_result
```
